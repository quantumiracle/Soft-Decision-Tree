```python
import os
import numpy as np
directory='state/'
s_data=[]

for filename in os.listdir(directory):
    if filename.endswith(".npy"): 
        filepath = directory+filename
        s = np.load(filepath)
        s_data.append(s)

data_size = len(s_data)
s_shape = s_data[0].shape
print(data_size, s_shape)
```

```python
import matplotlib.pyplot as plt

n_channel = s_shape[0]
f, ax = plt.subplots(1,n_channel, figsize=(20,5))
print(n_channel)

for i in range(n_channel):
    ax[i].imshow(s_data[1][i])
plt.show()
```
